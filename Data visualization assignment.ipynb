```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from sklearn import decomposition
from sklearn import datasets
```

```python
iris =sns.load_dataset('iris')
iris.shape
```

```python
x=iris[['sepal_length','sepal_width','petal_length','petal_width']]
y=iris['species']
```

```python
s=np.array(iris,ndmin=3)
```

```python
pca =decomposition.PCA(n_components=3)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents,columns = ['pc1', 'pc2','pc3'])

finalDf= pd.concat([principalDf, iris[['species']]], axis = 1)
finalDf
```

```python
fig = plot.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection='3d')
targets = ['setosa', 'versicolor', 'virginica']
colors = ['blue', 'yellow', 'orange']
for target, color in zip(targets,colors):
    indices = finalDf['species'] == target
    #print(indicesToKeep,target,color)
    ax.scatter(finalDf.loc[indices,'pc1'], finalDf.loc[indices,'pc2'], finalDf.loc[indices,'pc3'], c = color, s = 50)
    #break
    
ax.legend(targets)
plot.show()
```
